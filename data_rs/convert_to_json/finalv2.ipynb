```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
PATH = 'gdrive/MyDrive/proyekBangkit/finalv2.csv'
df = pd.read_csv(PATH)
```

```python
df.head()
```

```python
df.drop(['Unnamed: 0', 'provinsi'], axis=1, inplace=True)
```

```python
df.head()
```

```python
rating_mean = df['rating_average'].mean()
rating_90th = df['rating_count'].quantile(0.90)
# rating_90th = df['rating_count']
print(rating_90th)
```

```python
# new_df = df.copy().loc[df['rating_count'] >= rating_90th]
```

```python
def weighted_rating(df, quantil=rating_90th, mean=rating_mean):
    rating_count_series = df['rating_count']
    rating_average_series = df['rating_average']

    return (rating_count_series / (rating_count_series + quantil) * rating_average_series) + (quantil / (quantil + rating_count_series) * mean)
```

```python
df['score'] = df.apply(weighted_rating, axis=1)
```

```python
df = df.sort_values('score', ascending=False)
```
